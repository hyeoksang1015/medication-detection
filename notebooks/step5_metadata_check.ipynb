```python
import json
import os
from collections import Counter

merged_path = os.path.join(os.getcwd(), "merged_annotations_clean.json")
with open(merged_path, "r", encoding="utf-8") as fp:
    coco = json.load(fp)

light_color_counts = Counter(img.get("light_color", "N/A") for img in coco["images"])
back_color_counts = Counter(img.get("back_color", "N/A") for img in coco["images"])

print(f"전체 이미지 개수: {len(coco['images'])}")

print("\n=== light_color 분포 ===")
for color, count in light_color_counts.most_common():
    print(f"  {color}: {count}개")

print("\n=== back_color 분포 ===")
for color, count in back_color_counts.most_common():
    print(f"  {color}: {count}개")

combo_counts = Counter(
    (img.get("light_color", "N/A"), img.get("back_color", "N/A"))
    for img in coco["images"]
)

print("\n=== light_color x back_color 조합 분포 ===")
for (light, back), count in combo_counts.most_common():
    print(f"  light={light}, back={back}: {count}개")

camera_la_counts = Counter(img.get("camera_la", "N/A") for img in coco["images"])
print("\n=== camera_la (각도) 분포 ===")
for angle, count in camera_la_counts.most_common():
    print(f"  {angle}도: {count}개")
```
