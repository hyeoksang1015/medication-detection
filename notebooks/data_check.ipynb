```python
from pathlib import Path

def find_project_root() -> Path:
    for candidate in [Path.cwd(), *Path.cwd().parents]:
        target = candidate / "data" / "raw"
        print("확인 중:", target, "=>", target.is_dir())

        if target.is_dir():
            print("프로젝트 루트 찾음:", candidate)
            return candidate

    raise FileNotFoundError("data/raw 폴더를 찾지 못함")

ROOT        = find_project_root()
DATA_RAW    = ROOT / "data" / "raw"
DATA_PROC   = ROOT / "data" / "processed"

KAGGLE_IMG  = DATA_RAW / "kaggle_data" / "train_images"
KAGGLE_TEST = DATA_RAW / "kaggle_data" / "test_images"
AIHUB_IMG   = DATA_RAW / "aihub_data" / "aihub_images"
AIHUB_ANN   = DATA_RAW / "aihub_data" / "aihub_annotations"

YOLO_DIR    = ROOT / "data" / "yolo"
MODELS_DIR  = ROOT / "models"

DATA_PROC.mkdir(parents=True, exist_ok=True)
YOLO_DIR.mkdir(parents=True, exist_ok=True)
MODELS_DIR.mkdir(parents=True, exist_ok=True)

print("ROOT       =", ROOT)
print("DATA_RAW   =", DATA_RAW)
print("AIHUB_IMG  =", AIHUB_IMG)
print("AIHUB_ANN  =", AIHUB_ANN)
```
