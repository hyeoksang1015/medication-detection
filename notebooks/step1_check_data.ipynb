```python
import json
import os

base_path = r"C:\Users\Administrator\.cache\kagglehub\competitions\ai12-level1-project"

image_count = 0
for root, dirs, files in os.walk(base_path):
    if "train_images" in root:
        image_count += len([f for f in files if f.endswith((".png", ".jpg", ".jpeg"))])

print(f"train_images 총 이미지 개수: {image_count}")

json_count = 0
for root, dirs, files in os.walk(base_path):
    json_count += len([f for f in files if f.endswith(".json")])

print(f"전체 JSON 파일 개수: {json_count}")

sample_json = None
for root, dirs, files in os.walk(base_path):
    for f in files:
        if f.endswith(".json"):
            sample_json = os.path.join(root, f)
            break
    if sample_json:
        break

if sample_json:
    print(f"샘플 JSON 파일: {sample_json}")
    with open(sample_json, "r", encoding="utf-8") as fp:
        data = json.load(fp)
    print("JSON 최상위 키:", list(data.keys()))
    print(json.dumps(data, indent=2, ensure_ascii=False)[:1000])
```
